```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("02 modern-renewable-energy-consumption.csv")
df
```

I am removing the entries that are listed as world. That way, we get a better understanding of induvial countries and international organizations. 

```python
df_entity = df[df['Entity'] != 'World']
df_entity
```

Here is the data regarding only consumption of geo biomass. This is the energy generated from living/once-alive organisms such as plants and animals. 

```python
df_Geo_Biomass = df_entity.iloc[:, [0, 2, 3]]
df_Geo_Biomass
```

Here is the data regarding only solar energy generation. This is the energy generated from the sun through methods like solar panels. 

```python
df_Solar = df_entity.iloc[:, [0, 2, 4]]
df_Solar
```

Here is the data regarding only wind energy generation. This is the energy generated harnessing the wind, through things like wind turbines. 

```python
df_Wind = df_entity.iloc[:, [0, 2, 5]]
df_Wind
```

Here is the data regarding only hydro energy generation. This is the energy generated from harnessing water and its movements. 

```python
df_Hydro = df_entity.iloc[:, [0, 2, 6]]
df_Hydro
```

Here I isolated just the data regarding the solely U.S. 

```python
df_US = df[df['Entity'] == 'United States']
df_US
```

**Here are visual representations of the United States in regards to their usage of renewable energy:**

```python
df_US.plot(x='Year', y='Geo Biomass Other - TWh', kind='line', title='United States Geo Biomass Consumption')
plt.show()
```

```python
df_US.plot(x='Year', y='Solar Generation - TWh', kind='line', title='United States Solar Energy Generation from 1965 to 2021')
plt.show()
```

```python
df_US.plot(x='Year', y='Wind Generation - TWh', kind='line', title='United States Wind Energy Generation from 1965 to 2021')
plt.show()
```

```python
df_US.plot(x='Year', y='Hydro Generation - TWh', kind='line', title='United States Hydro Energy Generation from 1965 to 2021')
plt.show()
```

```python
df_US.plot(x = 'Year', kind='area', title='United States Renewable Energy Practices from 1965 to 2021')
plt.show()
```

**Here is just the data regarding the year 2021:**

```python
df_2021 = df_entity[df_entity['Year'] == 2021]
df_2021
```

Here is the data from 2021 showing who are the top 20 global entities tht have used renewble energies. 

```python
df_largest_Biomass = df_2021.nlargest(n=20, columns='Geo Biomass Other - TWh')
values = df_largest_Biomass['Geo Biomass Other - TWh']
labels = df_largest_Biomass['Entity']
plt.pie(values, labels=labels)
plt.title('Top 20 Geo Biomass Users')
plt.show()
```

```python
df_largest_Solar = df_2021.nlargest(n=20, columns='Solar Generation - TWh')
values = df_largest_Solar['Solar Generation - TWh']
labels = df_largest_Solar['Entity']
plt.pie(values, labels=labels)
plt.title('Top 20 Solar Energy Users')
plt.show()
```

```python
df_largest_Wind = df_2021.nlargest(n=20, columns='Wind Generation - TWh')
values = df_largest_Wind['Wind Generation - TWh']
labels = df_largest_Wind['Entity']
plt.pie(values, labels=labels)
plt.title('Top 20 Wind Energy Users')
plt.show()
```

```python
df_largest_Hydro = df_2021.nlargest(n=20, columns='Hydro Generation - TWh')
values = df_largest_Hydro['Hydro Generation - TWh']
labels = df_largest_Hydro['Entity']
plt.pie(values, labels=labels)
plt.title('Top 20 Hydro Energy Users')
plt.show()
```
